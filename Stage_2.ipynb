```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('credits.csv')
credits.head()
```

```python
meta = pd.read_csv('movies_metadata.csv')
meta.head()
```

```python
meta.columns
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce')
```

```python
meta['year'] = meta['release_date'].dt.year
```

```python
meta['year'] = meta['release_date'].dt.year
meta['year'].head()
```

```python
meta['year'].value_counts().sort_index()
```

```python
# Getting only 2017 movies as we already have movies up to the year 2016 in preprocessing 1 file. 
# We don't have enough data for the movies from 2018, 2019 and 2020. 
# We'll deal with it in the upcoming preprocessing files
new_meta = meta.loc[meta.year == 2017,['genres','id','title','year']]
new_meta.head()
```

```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
data = pd.merge(new_meta, credits, on='id')
```

```python
pd.set_option('display.max_colwidth', 75)
data
```

```python
# evaluates an expression node or a string containing a Python literal or container display
import ast
```

```python
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data['genres'][26]
```

```python
def genreList(value):
    listValue = []
    string  = ' '
    for i in value:
        if i.get('name') == 'Science Fiction':
            temp = 'Sci-Fi'
            listValue.append(temp)
        else:
            listValue.append(i.get('name'))
    if listValue == []:
        return np.nan
    else:
        return string.join(listValue)
```

```python
data['genres_list'] = data['genres'].map(lambda x: genreList(x))
```

```python
data['genres_list']
```

```python
data['cast'][0]
```

```python
def castList1(value):
    listValue = []
    for i in value:
        listValue.append(i.get('name'))            
    if listValue == []:
        return np.nan
    else:
        return listValue[0]
```

```python
data['actor_1_name'] = data['cast'].map(lambda x : castList1(x))
```

```python
data['actor_1_name']
```

```python
def castList2(value):
    listValue = []
    for i in value:
        listValue.append(i.get('name'))            
    if listValue == [] or len(listValue) <= 1 :
        return np.nan
    else:
        return listValue[1]
```

```python
data['actor_2_name'] = data['cast'].map(lambda x : castList2(x))
```

```python
data['actor_2_name']
```

```python
def castList3(value):
    listValue = []
    for i in value:
        listValue.append(i.get('name'))            
    if listValue == [] or len(listValue) <= 2 :
        return np.nan
    else:
        return listValue[2]
```

```python
data['actor_3_name'] = data['cast'].map(lambda x : castList3(x))
```

```python
data['actor_3_name']
```

```python
def directorList(value):
    listValue = []
    st = ' '
    for i in value:
        if i.get('job') == 'Director':
            listValue.append(i.get('name'))            
    if listValue == []:
        return np.nan
    else:
        return st.join((listValue))
```

```python
data['crew'][0]
```

```python
data['director_name'] = data['crew'].map(lambda x: directorList(x))
data['director_name']
```

```python
movie = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
movie
```

```python
movie.isna().sum()
```

```python
movie = movie.dropna(how='any')
```

```python
movie.isna().sum()
```

```python
movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})
```

```python
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
```
